def correlation_ratio(cat_var, num_var):

    groups = [num_var[cat_var == c] for c in np.unique(cat_var)]
    ss_between = sum(len(g) * (g.mean() - num_var.mean()) ** 2 for g in groups)
    ss_total = sum((num_var - num_var.mean()) ** 2)
    return np.sqrt(ss_between / ss_total)

    
print(correlation_ratio(new_df['брак'], new_df['ингредиент_1']).round(4))
